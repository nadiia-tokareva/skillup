```python
# ==================================
# =         task №1                =
# ==================================
from pprint import pprint

# словарь, с данными игроков
basketball_players = {
    'Michael Jordan' : 198,
    'LeBron James' : 206,
    'Kareem Abdul-Jabbar' : 216,
    'Earvin Effay Johnson' : 206,
    "Shaquille Rashaun O'Neal" : 216
    
}

pprint(basketball_players)

# поиск информации
pprint(basketball_players['Michael Jordan'])

pprint(basketball_players.get('Johnson'))

pprint(basketball_players.get('LeBron James'))

pprint('Michael Jordan' in basketball_players)

# добавление игрока
new_basketball_players = {'Kobe Bean Bryant' : 198}
basketball_players.update(new_basketball_players)

pprint(basketball_players)

# удаление игрока
del basketball_players['LeBron James']

pprint(basketball_players)

# замена информации
basketball_players['LeBron James'] = 203

pprint(basketball_players)

```

```python
#создать англо-француский словарь

from pprint import pprint

vocabulary = {
    'morning' : 'matin',
    'evening' : 'soirée',
    'night' : 'nuit',
    'hello' : 'bonjour',
    'good bay' : 'bonne baie'
}

pprint(vocabulary)

# поиск информации
pprint(vocabulary.get('morning'))

pprint('apple' in vocabulary)

pprint('ball' not in vocabulary)

pprint(vocabulary['night'])

# добавление информации
new_words = {
    'man' : 'homme',
    'woman' : 'femme',
    'child' : 'enfant'
}

vocabulary.update(new_words)

pprint(vocabulary)

# удаление значения
vocabulary.pop('child')
del vocabulary['woman']

pprint(vocabulary)

# замена информации
vocabulary['morning'] = 'le matin'

pprint(vocabulary)
```

```python
# ==================================
# =         task №2                =
# ==================================

# Создать множества: A, B, C з любыми элементами.

A = {12, 'apple', True}
B = {'b', 5977779, (1,2,3), 'one'}
C = {'corn', 656565, False, 12}

print(A)
print(B)
print(C)

# Одинаковые элементы для A и C.
print(A.isdisjoint(C))   # содержат одинаковых значений
print(A.intersection(C))

# Разные элементы для A и B.
print(A.difference(B))

# Объединение 3-х множеств.
print(A.union(B,C))
```

```python
# ==================================
# =         task №3                =
# ==================================

a = (1, 2, 5, 9)
b = (9, 5, 0)
c = (6, 8, 9, 7)

print(a, b, c)

# найти элементы, которые есть во всех кортежах

for i in a:
    for x in b:
        for y in c:
            if i==x and x==y:
                print(i)
                
# найти элементы, которые уникальны для каждого списка

unique_a = (set(a)-set(b)-set(c))
unique_b = (set(b)-set(c)-set(a))
unique_c = (set(c)-set(a)-set(b))

print(unique_a)
print(unique_b)
print(unique_c)
```
